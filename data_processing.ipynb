## Date Processing

```python
import pandas as pd
import re
```

### Clean Wine Data

```python
df = pd.read_csv("asset/wines.csv")
df
```

```python
df['flavors'] = df['flavors'].apply(lambda x: dict(zip(re.findall('\'group\W*(.*?)\W', x), re.findall('stats\W*\'count\W*(.*?)\W', x))))
df
```

### Clean Review Data

```python
df2 = pd.read_pickle("asset/reviews.pkl")
df2
```

```python
df2.drop('flavors', axis=1, inplace=True)
df2 = df2.merge(df[['wine ID', 'year', 'flavors']], how='left', on=['wine ID', 'year'])
df2
```

### Clean Country Data

```python
df3 = pd.read_csv("asset/country.csv")
df3
```

```python
grapes = df3['most_used_grapes'].apply(lambda x: re.findall('{(.*?)}', x))
split_grapes = pd.DataFrame(grapes.tolist(), columns=['grape_1', 'grape_2', 'grape_3'])
split_grapes.fillna(value='no values', inplace=True)


split_grapes['grape_1_id'] = split_grapes['grape_1'].apply(lambda x: re.findall('id\W*(.*?)\W', x))
split_grapes['grape_1_id'] = split_grapes['grape_1_id'].apply(lambda x: x[0] if len(x) > 0 else None)

split_grapes['grape_1_name'] = split_grapes['grape_1'].apply(lambda x: re.findall('\'name\W*(.*?)\W', x))
split_grapes['grape_1_name'] = split_grapes['grape_1_name'].apply(lambda x: x[0] if len(x) > 0 else None)

split_grapes['grape_2_id'] = split_grapes['grape_2'].apply(lambda x: re.findall('id\W*(.*?)\W', x))
split_grapes['grape_2_id'] = split_grapes['grape_2_id'].apply(lambda x: x[0] if len(x) > 0 else None)

split_grapes['grape_2_name'] = split_grapes['grape_2'].apply(lambda x: re.findall('\'name\W*(.*?)\W', x))
split_grapes['grape_2_name'] = split_grapes['grape_2_name'].apply(lambda x: x[0] if len(x) > 0 else None)

split_grapes['grape_3_id'] = split_grapes['grape_3'].apply(lambda x: re.findall('id\W*(.*?)\W', x))
split_grapes['grape_3_id'] = split_grapes['grape_3_id'].apply(lambda x: x[0] if len(x) > 0 else None)

split_grapes['grape_3_name'] = split_grapes['grape_3'].apply(lambda x: re.findall('\'name\W*(.*?)\W', x))
split_grapes['grape_3_name'] = split_grapes['grape_3_name'].apply(lambda x: x[0] if len(x) > 0 else None)

split_grapes.drop(['grape_1', 'grape_2', 'grape_3'], axis=1, inplace=True)

split_grapes
```

```python
df3 = df3.join(split_grapes)
df3.drop('most_used_grapes', axis=1, inplace=True)
df3
```
